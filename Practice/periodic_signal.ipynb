### 사인 곡선 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 10, 0.1)
amplitude = np.sin(t)

plt.plot(t, amplitude)

plt.title('Sine Wave')
plt.xlabel('time')
plt.ylabel('amplitude = sin(time)')
plt.grid(True, which='both' )
plt.axhline(y=0, color='k') # x축을 검은색으로 찐하게
plt.axvline(x=0, color='k')

plt.show()
```

### 여러 사인곡선이 합해진 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 30, 0.1)
g1 = np.sin(t)
g2 = np.sin(2*t)
g3 = np.sin(3*t)

g = g1 + g2 + g3

plt.plot(t, g)

plt.show()
```

### 펄스파 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

t = np.linspace(0, 1, 1000)
plt.plot(t, signal.square(2*np.pi*5*t))

# 주파수 5Hz의 사각파를 생성
# 2*np.pi*5는 사각파의 각 주파수를 라디안 단위로 변환
```

### 라디안값으로 사인 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
start = -2
end = 4

x = np.arange(start*pi, end*pi+0.1, 0.1)
y = np.sin(x)

plt.plot(x,y)
ticks = np.arange(start*pi,end*pi+0.1,pi)
labels = []
for i in range(start, end+1):
    if i == 0: labels.append('0')
    elif i==-1: labels.append('-π')
    elif i==1: labels.append('π')
    else: labels.append(str(i) + 'π')

plt.xticks(ticks = ticks, labels=labels)
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k')
plt.grid(True, which='major', axis='x' )
```
