```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일을 읽기
df = pd.read_csv('./eval_score.csv')

# 파일 이름에서 카테고리 추출 (예: eval_ball, eval_inner 등)
df['Category'] = df['File'].apply(lambda x: x.split('_')[1])

# 카테고리별로 히스토그램 그리기
categories = ['ball', 'inner', 'normal', 'outer']

plt.figure(figsize=(10, 6))

for category in categories:
    subset = df[df['Category'] == category]
    plt.hist(subset['Score'], bins=10, alpha=0.5, label=f'eval_{category}')

plt.title('Score Distribution by Category')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```
